```python
from pyspark.sql.functions import current_timestamp, lit, input_file_name

tables = ["customers", "exchange_rates","products","sales","stores"]
 
for tbl in tables:
    df = spark.table(f"electronic_retailer.azure_blob_storage.{tbl}") \
              .withColumn("_ingested_at", current_timestamp()) \
              .withColumn("_source_system", lit("fivetran"))
    df.write.mode("overwrite").saveAsTable(f"electronic_retailer.01_Bronze.{tbl}_raw")
    print(f"Bronze: {tbl}_raw loaded with the {df.count()} rows")
```
